1. **Figure 91 - Group 1 (Arin and Engin)**:

2. **Figure 93 - Group 1 (Cello and Ezgi)**:  
   A steady increase in MLS alongside a rising **CN/C** demonstrates consistent growth in both fluency and complexity.

3. **Figure 96 - Group 1 (Helin and Senay)**:  
   This graph shows moderate progress in both measures, with a smoother upward trend in **CN/C** compared to MLS, indicating syntactic complexity is developing slightly faster than sentence length.

4. **Figure 99 - Group 1 (Selin and Zahide)**:  
   A sharp rise in **CN/C** with a more gradual increase in MLS suggests that while sentence length is improving, syntactic structures are becoming much more complex.
